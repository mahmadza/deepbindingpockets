```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(1)

X_xor = np.random.randn(200,2)
X_xor
```

```python
y_xor = np.logical_xor(X_xor[:, 0] > 0,
                      X_xor[:, 1] > 0)
y_xor = np.where(y_xor, 1, -1)
y_xor
```

```python
plt.scatter(X_xor[y_xor == 1, 0],
           X_xor[y_xor ==1, 1],
           c = 'b', marker = 'x',
           label = 1)
plt.scatter(X_xor[y_xor == -1, 0],
           X_xor[y_xor == -1, 1],
           c = 'r', marker = 's',
           label = -1)
plt.xlim([-4, 4])
plt.ylim([-4, 4])
plt.legend(loc = 'best')
plt.show()
```
